```python
import pandas as pd
import matplotlib.pyplot as plt  
```

## Analyse d'un dataset

```python
dataset = pd.read_csv(r'/Users/sabrigasmi/Desktop/Dataset/2019.csv')
# Possibilité de changer le paramètre par pearson par exemple
dataset.corr("spearman")
```

## Création d'un nuage de point avec matplotlib, avec pyplot

```python
# Nuage de points, lien entre GDP pet capita and Social support

plt.figure(figsize=(12,8))
x = dataset["GDP per capita"]
y = dataset["Healthy life expectancy"]

plt.xlabel("GDP per capita")
plt.ylabel("Healthy life expectancy")

plt.title("Link between GDP and Healhty life expectancy")
plt.scatter(x,y, label="Country or region")
plt.legend()
```

```python
# Nuage de points avec legende, lien entre GDP pet capita and Social support
plt.figure(figsize=(17,10))
x = dataset["Social support"]
y = dataset["GDP per capita"]

# Ajout des labels sur les points (ici trop de points pour être lisible)

for i, val in enumerate(dataset.iterrows()):
    plt.annotate(dataset.at[i,"Country or region"], # this is the text
                 (dataset.at[i, "Social support"],dataset.at[i, "GDP per capita"]), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
    
plt.scatter(x,y)
plt.savefig('mafigure.png', dpi=300)
```

## Création d'un graphique avec subplot

```python
plt.figure(figsize=(12,8))

# Utilisation de subplot pour créer plusieurs graphique dans un graphique

# Premier paramètre les lignes, puis les colonnes, enfin le graphique auquel on fait référence


plt.subplot(2,1,1)
plt.scatter(dataset["GDP per capita"], dataset["Healthy life expectancy"],c="red")
plt.xlabel("GDP per capita")
plt.ylabel("Healthy life expectancy")
plt.title("Study of Healthy life expectancy")

plt.subplot(2,1,2)
plt.scatter(dataset["GDP per capita"], dataset["Perceptions of corruption"],c="blue")
plt.xlabel("GDP per capita")
plt.ylabel("Perceptions of corruption")
plt.title("Study of perceptions of corruption")

plt.tight_layout(pad=3.0) # Espacement entre graphique
```
